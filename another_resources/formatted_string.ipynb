## str.format()

```python
"""
str.format() = optional method that gives users **more control when displaying output**
"""

animal = 'cat'
item = 'sun'

# Understand the basic
print('The {} jumped over the {}'.format('cat','sun')) # u can fill it with string or any data u want
print('The {} jumped over the {}\n'.format(animal,item)) # u can even fill it with variable

# Whoa index?
print('The {0} jumped over the {1}'.format(animal,item)) # really? there's index in str.format?!
print('The {0} jumped over the {0}'.format(animal,item)) # owh im starting to understand...
print('The {1} jumped over the {0}'.format(animal,item)) # IT'S POSITIONAL ARGUMENT
print('The {animal} jumped over the {item}\n'.format(animal='crocodile',item='building')) # yup keyword argument

# flexibility
text = 'The {} jumped over the {}'
print(text.format(animal,item),'\n') # nice flexibility
```

```python
# more complex 🤙
name = 'Fulan'

print('Hello, my name is {}. Nice to meet u'.format(name)) # pfft boring
print('Hello, my name is {:10}. Nice to meet u'.format(name)) # wait what happen? (automatic left for str)
print('Hello, my name is {:>10}. Nice to meet u'.format(name)) # right alignment
print('Hello, my name is {:^10}. Nice to meet u'.format(name)) # center alignment
print('Hello, my name is {:<10}. Nice to meet u\n'.format(name)) # left alignment

# for numbers
number = 3.14159

print('The number pi is {}'.format(number)) # i know this
print('The number pi is {:10}'.format(number)) # display n space for value (automatic right for number)
print('The number pi is {:.2}'.format(number)) # display n part of the value (somehow it only works for float, idk)
print('The number pi is {:.2f}'.format(number)) # display n decimal point (f stands for floating number!)
print('The number is {:,}'.format(9999999999)) # separate number with comma per thousand
print('The number is {:b}'.format(200)) # display number as binary digit
print('The number is {:o}'.format(200)) # display number as octal digit
print('The number is {:d}'.format(200)) # display number as decimal digit (standard)
print('The number is {:x}'.format(200)) # display number as hexadecimal digit (lowercase)
print('The number is {:X}'.format(200)) # display number as hexadecimal digit (uppercase)
print('The number is {:e}'.format(200)) # display number as scientific number (lowercase)
print('The number is {:E}'.format(200)) # display number as scientific number (uppercase)
print('The number pi is {:%}'.format(number)) # converts numbers to percentages and automatically adds the % symbol
print('The number pi is {:20%}'.format(number)) # determine its spaces after converts to percentages
print('The number pi is {:.2%}\n'.format(number)) # determine its decimal point after converts to percentages

# fun fact: when we decide to specify spaces for our value, number and string reacts differently
print('test {:20,} test'.format(100000000)) # automatic justify to right
print('test {:20} test\n'.format('hello world')) # automatic justify to left

# idk the caption
print('{:#^20}'.format('SUCCESS'))
print('{:_^+20}'.format(100))
print('{:_^ 20}'.format(100))
```

## f-string

```python
"""
format specifiers = {value:flags} format a value based on what flags are inserted
basically the same as str.format(),f-string just the modern verse of it
personally i love to use f-string more than str.format()
"""

a = 'Hello'
b = 'World!'

print(f'{a} {b}')
print(f'{a:=^20} {b:=>20}\n')

price1 = 30000.14156
price2 = -9851.131
price3 = 1200.42

print(f'Price 1 is ${price1: ,.2f}')
print(f'Price 2 is ${price2: ,.2f}')
print(f'Price 3 is ${price3: ,.2f}')
```
